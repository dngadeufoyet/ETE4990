````python
# Define a Player class to represent a player in the game.
class Player:
    # The constructor initializes the player with a name and empty inventory and team lists.
    def __init__(self, name):
        self.name = name
        self.inventory = []  # Stores the player's items
        self.team = []  # Stores the player's Pokemon

    # Method to add an item to the player's inventory
    def add_item(self, item):
        self.inventory.append(item)
        
    # Method to remove an item from the player's inventory
    def remove_item(self, item):
        self.inventory.remove(item)
        
    # Method to display all items in the player's inventory
    def display_inventory(self):
        print(f"{self.name}'s Inventory:")
        for item in self.inventory:
            print(f"{item['name']} - {item['description']}")
        
    # Method to add a Pokemon to the player's team
    def add_pokemon(self, pokemon):
        self.team.append(pokemon)

    # Method to display all Pokemon in the player's team
    def display_team(self):
        print(f"{self.name}'s Team:")
        for pokemon in self.team:
            print(f"{pokemon.name} - HP: {pokemon.health}/{pokemon.max_health}")

# Define a Pokemon class to represent a Pokemon in the game.
class Pokemon:
    # The constructor initializes the Pokemon with a name, type, and health.
    def __init__(self, name, type, health):
        self.name = name
        self.type = type
        self.health = health
        self.max_health = health  # Max health is set to the initial health value
        self.moves = []  # Stores the Pokemon's moves

    # Method to add a move to the Pokemon's moveset
    def add_move(self, move):
        self.moves.append(move)
        
    # Method to display all moves of the Pokemon
    def display_moves(self):
        print(f"{self.name}'s Moves:")
        for move in self.moves:
            print(f"{move['name']} - Damage: {move['damage']}%")

# Example items and pokemon creation

# Define a pokeball item with a description and a trait
pokeball = {
    "name": "Pokeball",
    "description": "A device for catching wild Pokemon.",
    "trait": "Catch rate 30%",
}

# Define a swordfish item with a description and a trait
swordfish = {
    "name": "Aqua Blade",
    "description": "A device for killing wild Pokemon.",
    "trait": "Catch rate 60%",
}

# Define an elixir item with a description and a trait
elixir_of_life = {
    "name": "Elixir of Life",
    "description": "Restores all health.",
    "trait": "Healing item",
}

elixir_of_power = {
    "name": "Elixir of power",
    "description": "Increase your power",
    "trait": "Powerful FireBall",
}

# Create a Pokemon instance 'Squirtle' with a type and health
squirtle = Pokemon("Squirtle", "Water", 75)
# Add a move 'Aqua Jet' to Squirtle's moveset
squirtle.add_move({"name": "Aqua Jet", "damage": 35})

# Create a Water-type Pokemon instance 'Swordfish' with a type and health
swordfish_pokemon = Pokemon("Swordfish", "Water", 100)
# Add a move 'Aqua Blade' to Swordfish's moveset with power 80
swordfish_pokemon.add_move({"name": "Aqua Blade", "damage": 80})

player = Player(input("Enter your player name: "))
player_type = input("Enter your player type (pokemon/swordfish): ").lower()
while player_type not in ["pokemon", "swordfish"]:
    print("Invalid player type.")
    player_type = input("Enter your player type (pokemon/swordfish): ").lower()

# Add items to player's inventory
player.add_item(pokeball)
player.add_item(elixir_of_life)

# Add a Pokemon to player's team
if player_type == "pokemon":
    player.add_pokemon(squirtle)
elif player_type == "swordfish":
    player.add_pokemon(swordfish_pokemon)

# Display all inventory 
player.display_inventory()
# Display player's team
player.display_team()

while True:
    print("\n1. Add a new avatar")
    print("2. End")
    choice = input("Enter your choice: ")
    if choice == "1":
        player = Player(input("Enter your player name: "))
        player_type = input("Enter your player type (pokemon/swordfish): ").lower()
        while player_type not in ["pokemon", "swordfish"]:
            print("Invalid player type.")
            player_type = input("Enter your player type (pokemon/swordfish): ").lower()
        player.add_item(elixir_of_power)
        if player_type == "pokemon":
            player.add_pokemon(squirtle)
        elif player_type == "swordfish":
            player.add_pokemon(swordfish_pokemon)
        player.display_inventory()
        player.display_team()
    elif choice == "2":
        print("Ending the program.")
        break
    else:
        print("Invalid choice. Please try again.")
````
